# Unsupervised Machine Learning 

```python
from sklearn.datasets import load_wine

wine = load_wine()
print(wine.DESCR)
```

```python
print(f'wine.data.shape    = {wine.data.shape}')
print(f'wine.target.shape  = {wine.target.shape}')
print(f'wine.target_names  = {wine.target_names}')
print(f'wine.feature_names = {wine.feature_names}')
```

```python
print(wine.data)
```

```python
import pandas as pd

pd.set_option('max_columns', 14)
pd.set_option('display.width', None)

wine_df = pd.DataFrame(wine.data, columns=wine.feature_names)
wine_df['Type'] = [wine.target_names[i] for i in wine.target]

wine_df.head(5)
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=3, random_state=11)  # find three clusters
kmeans.fit(wine.data)
```

# Dimensionality reduction with TSNE

```python
from sklearn.manifold import TSNE

tsne = TSNE(n_components=2, random_state=11)
reduced_wine = tsne.fit_transform(wine.data)

print(f'Reduced data shape: reduced_wine.shape = {reduced_wine.shape}')
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

dots = plt.scatter(reduced_wine[:, 0], reduced_wine[:, 1], c='black')
```

```python
dots = plt.scatter(reduced_wine[:, 0], reduced_wine[:, 1], c=wine.target, 
                   cmap=plt.cm.get_cmap('nipy_spectral_r', 3))

colorbar = plt.colorbar(dots)
```

# Dimensionality reduction with Principal Component Analysis

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2, random_state=11)  # reduce to two components
pca.fit(wine.data)
```

```python
reduced_wine = pca.transform(wine.data)
reduced_wine.shape
```

```python
reduced_wine_df = pd.DataFrame(reduced_wine, 
                               columns=['Component 1', 'Component 2'])
reduced_wine_df['Type'] = wine_df.Type #Labels. type of flower in the original dataframe column
```

```python
import matplotlib.pyplot as plt

axes = sns.scatterplot(data=reduced_wine_df, hue='Type', legend='brief', 
                       x='Component 1', y='Component 2') 

wine_centers = pca.transform(kmeans.cluster_centers_)
dots = plt.scatter(wine_centers[:,0], wine_centers[:,1], s=100, c='k')
```

# Supervised Machine learning

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
     wine.data, wine.target, random_state=11)

print(f'X_train.shape = {X_train.shape}')
print(f'X_test.shape  = {X_test.shape}')

print()
print(f'y_train.shape = {y_train.shape}')
print(f'y_test.shape  = {y_test.shape}')
```

```python
wine.target
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X=X_train, y=y_train)
```

```python
predicted = knn.predict(X=X_test)
expected = y_test

print('First thirty predictions:')
print(f'predicted[:30] = {predicted[:30]}')   #Maximum is 42 beacuse it takes 25% in  sample for prediction.
print(f' expected[:30] = {expected[:30]}')
```

```python
wrong = [ (pred, exp) 
          for (pred, exp) in zip(predicted, expected) 
          if pred != exp
        ]

print('Wrong predictions:')
print(wrong)          #13 out of 42 have made wrong predictions.
```

```python
print(f'Prediction accuracy score = {knn.score(X_test, y_test):.2%}')
```

# Confusion Matrix

```python
from sklearn.metrics import confusion_matrix

confusion = confusion_matrix(y_true=expected, y_pred=predicted)

print('Confusion matrix:')
print(confusion)
```

```python
import pandas as pd
import seaborn as sns

confusion_df = pd.DataFrame(confusion, index=range(3), columns=range(3))
axes = sns.heatmap(confusion_df, annot=True, cmap='nipy_spectral_r')
```

# Classification Report

```python
from sklearn.metrics import classification_report

names = [str(wine) for wine in wine.target_names]
print(classification_report(expected, predicted, target_names=names))
```
